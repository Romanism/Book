### 1.1 왜 머신러닝인가?

: 머신러닝은 한분야의 데이터가 아니라 여러분야의 데이터를 학습할 수 있다

#### 1.1.1 머신러닝으로 풀 수 있는 문제

- 지도학습(supervised learning) 
    - $X$(입력값), $y$(출력값)을 통해 학습한다. 즉 정답이 있는 데이터를 토대로 학습하는 방법이다
    - ex) 편지봉투에 손으로 쓴 우편번호 숫자 판별, 의료영상 이미지에 기반한 종양판단


- 비지도학습(unsupervised learning)
    - $X$(입력값) 만을 통해 학습하는 방법이다
    - ex) 블로그 글의 주제 구분, 고객들을 취향이 비슷한 그룹으로 묶기

```python
# DataFrame
import pandas as pd

data = {'Name': ["John", "Anna", "Peter", "Linda"],
        'Location' : ["New York", "Paris", "Berlin", "London"],
        'Age' : [24, 13, 53, 33]
       }

data_pandas = pd.DataFrame(data)
data_pandas
```

- 샘플(Sample) : 행 (0, 24, New York, John)을 의미한다
- 특성(Feature) : 열 (Age, Location, Name)을 의미한다 

#### 1.1.2 문제와 데이터 이해하기

- 어떤 질문에 대한 답을 원하는가? 가지고 있는 데이터가 원하는 답을 줄 수 있는가?
- 내 질문을 머신러닝의 문제로 가장 잘 기술하는 방법은 무엇인가?
- 문제를 풀기에 충분한 데이터를 모았는가?
- 내가 추출한 데이터의 특성은 무엇이며 좋은 예측을 만들어 낼 수 있을 것인가?
- 머신러닝 애플리케이션의 성과를 어떻게 측정할 수 있을까?
- 머신러닝 솔루션이 다른 연구나 제품과 어떻게 협력할 수 있는가?

#### 1.7.2 성과 측정 : 훈련 데이터와 테스트 데이터

- 우리가 만든 모델은 새 데이터에 적용해봐야 진짜 모델의 예측성과 신뢰성을 알 수 있다
- 일반적으로 Train data (훈련데이터)로 모델을 만들고 Test data (테스트 데이터)로 모델 성능을 평가한다
- 따로 테스트 데이터가 없다면 트레인 데이터에서 일부를 떼내어 테스트 데이터로 이용한다

```python
# Train data로 Train/Test 만들기 (75% : 25%)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
# 원래 150개 데이터가 112/38로 분리된 것을 확인
print("X_train shape : {}".format(X_train.shape))
print("X_test shape : {}".format(X_test.shape))
print("y_train shape : {}".format(y_train.shape))
print("y_test shape : {}".format(y_test.shape))
```

#### 1.7.3 가장 먼저 할 일 : 데이터 살펴보기

```python
import mglearn
import matplotlib.pyplot as plt
iris_dataframe = pd.DataFrame(X_train, columns = iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c = y_train, figsize = (10, 10), marker = 'o',
                           hist_kwds = {'bins': 20}, s = 30, alpha = .8, cmap = mglearn.cm3)
plt.show()
```

#### 1.7.4 첫번째 머신러닝 모델 : K-최근접 이웃 알고리즘

- K : train 데이터에서 새로운 데이터 포인트에 가장 가까운 'k개'의 이웃을 찾는다는 의미이다
- 이웃들의 클래스 중 빈도가 가장 높은 클래스를 예측값으로 사용한다

```python
# KNeighborsClassifier 패키지는 train데이터를 자체적으로 저장하고 있음
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# model 만들기 (.fit)
model = knn.fit(X_train, y_train)
model
```

#### 1.7.5 예측하기

```python
# 예측하기 (model.predict)
prediction = model.predict(X_test) # train 데이터를 통해 모델을 생성, 모델을 통해 test데이터의 예측값을 생성
prediction
```

#### 1.7.6 모델 평가하기

```python
# 정확도 구하기1 (직접 비교)
print('정확도 : {:.2f}'.format(np.mean(prediction == y_test)))
```

```python
# 정확도 구하기2 (model.score)
print('정확도 : {:.2}'.format(model.score(X_test, y_test)))
```

### 1.8 요약 및 정리

1. 알고리즘을 통해 모델을 생성한다 (fit)
2. 생성한 모델을 바탕으로 테스트 데이터를 넣어본다 (predict)
3. 모델이 적합한지 평가한다 (score)

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

knn = KNeighborsClassifier(n_neighbors=1)
model = knn.fit(X_train, y_train)

print("정확도 : {:.2f}".format(model.score(X_test, y_test)))
```
